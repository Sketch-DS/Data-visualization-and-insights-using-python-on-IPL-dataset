```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl=pd.read_csv("ipl.csv")
ipl.head()
```

```python
ipl.shape
```

# PLayers with highest mom's

```python
ipl["player_of_match"].value_counts()
```

# Top 10 players with highest no. of player of the match

```python
ipl["player_of_match"].value_counts()[0:10]
```

# Bar-plot for highest no. of player of the match awards

```python
plt.bar(ipl["player_of_match"].value_counts()[0:5].keys(),ipl["player_of_match"].value_counts()[0:5],color="orange")
plt.xlabel("PLayer-Names")
plt.ylabel("Player of the matches frequency")
plt.title("Distribution of Players with Their winning frequencies")
plt.show()
```

# What is the more no. of ways to win(By Defending, By Chasing or Tie)

```python
ipl["result"].value_counts()
```

# pie chart for no. of ways to win(By Defending, By Chasing or Tie)

```python
plt.pie(ipl["result"].value_counts(),labels=ipl["result"].value_counts().keys(),autopct="0.1f%%")
plt.title("Pie-chart for ways of match ending")
plt.show()
```

# Which team wins the maximum toss in IPL

```python
ipl["toss_winner"].value_counts()[0:5]
```

```python
plt.figure(figsize=(12,5))
plt.bar(ipl["toss_winner"].value_counts().keys()[0:5],ipl["toss_winner"].value_counts()[0:5],color=["r","g","y","b","olive"])
plt.xlabel("team-Names")
plt.ylabel("Toss win frequency")
plt.title("Distribution of teams with Their toss-winning frequencies")
plt.show()
```

# Finding out top 5 cities have maximum no. of matches?

```python
ipl["city"].value_counts()[0:5]
```

```python
plt.figure(figsize=(8,3))
plt.bar(ipl["city"].value_counts().keys()[0:6],ipl["city"].value_counts()[0:6],color="olive")
plt.xlabel("City")
plt.ylabel("Match Count")
plt.title("City vs Match count")
plt.show()
```

# Which umpire was used in maximum  matches

```python
plt.figure(figsize=(9,5))
plt.bar(ipl["umpire1"].value_counts()[0:5].keys(),ipl["umpire1"].value_counts()[0:5],color="Red")
plt.xlabel("Umpire-Names")
plt.ylabel("Match Count")
plt.title("Highest no. of matches by umpires")
plt.show()
```

# Considering win margin frequencies

```python
plt.hist(ipl["result_margin"],bins=5,color="silver")
plt.xlabel("win-margin")
plt.ylabel("Match Count")
plt.title("Distribution of win-margin vs match count")
plt.show()
```

# Find the winning percentage of Team with batting first

```python
plt.figure(figsize=(9,11))
a=ipl[ipl["result_margin"] > 10]["winner"].value_counts()
b=a.to_numpy()
c=a.index.to_numpy()
plt.pie(b,labels=c,autopct="%0.1f%%")
plt.title("Winning percentage of teams with batting first")
plt.show()
```

```python
# Find the winning percentage of Team with bowling first
```

```python
plt.figure(figsize=(9,29))
x=ipl[ipl["result_margin"] <= 10]["winner"].value_counts()
w=x.to_numpy()
v=x.index.to_numpy()
plt.pie(w,labels=v,autopct="%0.1f%%")
plt.title("Winning percentage of teams with bowling first")
plt.show()
```
